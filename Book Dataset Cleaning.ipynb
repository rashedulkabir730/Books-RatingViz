### Adjust and change values for better visualizations 

```python
unique = df['language_code'].unique()
print(unique)
```

```python
# replace any value with en = english 
#replace spa = spanish 
#replace fre = french
#replace nl = Dutch
#replace ara = arabic

df['language_code'] = df['language_code'].replace({'en-GB': 'English', 'en-CA': 'English', 'en-US': 'English', 'fre' : 'French', 
                                        'nl' : 'Dutch', 'ara' : 'Arabic', 'eng' : 'English', 'spa' : 'Spanish'})
```

```python
unique = df['language_code'].unique()
print(unique)
```

```python
unique = df['genre'].unique()
print(unique)
```
